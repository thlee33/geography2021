## Anaconda에서 Geopandas 설치 및 실행에 문제가 있는 경우, 대체 분석환경으로서 구름(goorm) IDE를 활용하는 방법

## 1. 구름에 접속하여 회원가입 후, 구름IDE(goormide)에 접속

- goorm.io(https://www.goorm.io/)에 접속한 후 회원 가입    
 : Google 계정 등을 연동하여 생성할 수도 있음   

- 스크롤 다운하여 구름IDE(groomide)가 나타나면 "서비스 바로가기" 링크를 클릭

- groomide 서비스 페이지에서 "대시보드" 버튼 클릭  
- 대시보드 화면에서 "새 컨테이너" 버튼 클릭    
- 컨테이너 이름 등을 입력하고, 소프트웨어 스택에서 "Jupyter"를 선택   
- 템플릿, OS, 추가 모듈 등은 기본 상태로 두어도 무방   
- "생성 (Ctrl + M) 버튼을 클릭    
- 아래 화면과 같이 컨테이너가 생성되고, "실행" 버튼을 클릭하면 컨테이너가 실행됨  

- 아래 화면에서 ▶을 클릭하고 "new run jupyter notebook"을 클릭    
 : 또는 터미널에서 jupyter notebook을 입력하고 실행해도 됨 
- 화면의 URL(https://python20210420-ocuct.run.goorm.io/?token=goorm) 부분을 클릭
 : 또는 복사 아이콘을 클릭한 다음 웹브라우저 새 탭을 열고 URL 붙여넣기하여 접속해도 됨  

- 아래와 같이 Anaconda의 Jupyter Notebook과 동일한 주피터 환경이 보여짐  
- New에서 Python 3를 클릭하여 새 노트북을 생성  

- pandas는 기본적으로 설치되어 있어 바로 로딩하면 됨
