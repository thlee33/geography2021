## 1. 파이썬

- 파이썬은 프로그램 개발이 가능한 개발 언어입니다.  
파이썬은 개발이 쉽고, 다양한 운영환경에서 동작이 가능하며,   
특히 오픈소스로서 커뮤니티가 발달하면서 많이 활용되고 있습니다.  
   
     
- 더구나 AI 알고리즘들도 많이 사용되고 있는 Python 환경을 우선적으로 지원하고 있기 때문에  
데이터 분석 및 시각화에서 가장 많이 이용되고 있는 분석환경이 되었습니다.

## 2. Anaconda (아나콘다)

파이썬을 기반으로 개발/분석할 수 있는 다양한 통합개발(분석)환경들이 있습니다.  
이 통합개발환경을 IDE(Integrated Development Environment) 또는 IDLE라고도 합니다.  
파이썬에서 가장 널이 사용되는 IDE 중의 하나가 Anaconda 입니다.
  
  
- Anaconda 장점  
 : 무료(커뮤니티) 버전을 제공합니다.  
 : 설치가 쉽고 편리하면서 이후의 관리도 용이합니다.  
 : 파이썬은 물론 공통적으로 많이 사용되는 패키지를 포함하고 있습니다.  
  
    
- Anaconda 단점   
 : 주요 패키지를 포함하고 있어서 용량이 큰 편입니다.  

### 2.1 아나콘다에서 파이썬 정상 동작 확인 방법

- 윈도우 시작 메뉴 > Anaconda3 (64-bit) > Anaconda Prompt 클릭  
- python 입력하고 엔터키 누름 
- 아래 화면과 같이 >>>로 파이썬 코드를 입력할 수 있는 파이썬 환경이 나타나면 정상 동작되고 있는 상태입니다. 
- exit()를 입력하고 엔터키를 누르거나, 창을 닫습니다. 

## 3. 주피터 노트북 (Jupyter Notebook)

주피터 노트북은 웹브라우저 상에서 노트 형태로 파이썬 코드를 입력하고 실행할 수 있는 개발/분석 환경입니다.  
cell이라고 부르는 문단 형태의 영역에 코드(Code)를 입력하고, cell별로 실행할 수 있어서  
데이터 로딩 > 전처리 > 분석 > 시각화 등의 단계별 진행 결과를 확인하고,   
필요시 전 단계로 돌아가서 코드를 수정하여 재실행하는 등의 피드백이 용이하기 때문에 분석에 매우 적합한 환경입니다.

### 3.1 주피터 실행 및 종료

#### 3.1.1 실행

주피터를 실행하는 2가지 방법이 있습니다.
1. PPT에 설명된 방법처럼 시작메뉴의 "Jupyter Notebook" 메뉴를 클릭  
2. Anaconda Prompt를 먼저 실행한 후, 여기에서 jupyter notebook을 입력하고 엔터키를 누르는 방법  

주피터가 정상적으로 실행되면 기본 웹브라우저에 Jupyter 화면과 기본 경로가 보여집니다.     
이 때, 터미널 창 하나가 뜨게 됩니다. 주피터를 웹 환경에서 이용할 수 있는 웹서버 프로세스를 실행하면서  
주피터 환경에서의 사용 기록(log)를 기록하는 역할도 하기 때문에 주피터 종료 전까지 이 창을 닫으면 안됩니다.   
(윈도우창 최소화는 상관없습니다) 

웹브라우저 창을 닫거나 한 경우, 웹브라우저를 띄우고 주소창에 http://localhost:8888/ 를 입력하여 접속해도 됩니다. 

#### 3.1.2 종료

주피터를 웹브라우저에서 이용할 수 있도록 웹서버 프로세스가 실행되는 터미널 창을 닫거나,     
터미널 창에 포커스를 준 상태에서 ctrl + c를 누르면 주피터 노트북 웹 서버 프로세스가 종료됩니다.   


### 3.2 주피터 주요 이용법 및 단축키

- 주피터에서 New를 눌러 새 ipynb가 생성되면 제목이 Untitled로 되어 있습니다. 마우스로 클릭하면 이름을 바꿀 수 있습니다.    


- Esc + B : 현재 Cell 아래에 빈 Cell을 생성한다.   
- Esc + A : 현재 Cell 위에 빈 Cell을 생성한다.   
- ESC : Cell 편집모드에서 빠져나온다


- ↑ : 바로 위 Cell로 이동한다.  
- ↓ : 바로 아래 Cell로 이동한다.  


- Shift + Enter : 현재 Cell이 실행 되며 실행 커서가 다음 Cell로 넘어 간다.  
- Ctrl + Enter : 현재 Cell이 실행 되고, Focus는 그대로 현재 Cell에 있다.  
- cell 실행을 중지 : esc + i + i    


- Cell 복사 : c  
- Cell 붙여넣기 : v 


- H : 키보드 단축키들을 팝업창으로 보여줌  


저장은 자동으로 되나, 수동으로 원할 경우, ctrl + s 키를 누르면 됩니다.

### 3.3 마크다운 (Markdown)

또한, cell에 코드만 입력할 수 있는 것이 아니고 Markdown이라고 하는 Mark up 언어를 지원하기 때문에 문서화를 하기에도 용이합니다.  
지금 보고 계시는 화면도 주피터에서 Cell을 Markdown 형식으로 바꿔서 입력한 것입니다.  
주피터노트북의 기본 포맷은 ipynb 확장자를 가진 파일로 저장되어 Github 등을 통해 공유/배포가 쉽고  
파이썬 기본 포맷인 .py 및 .html, PDF 파일 등으로도 내보내기하여 활용할 수 있습니다. 

#### Markdown 기본 문법

- Cell를 Markdown으로 변환 : 빈 셀(cell)을 만든 후 M 키를 누르면 마크다운 형식 cell로 바뀝니다.     
  
  
- 줄바꿈 : 빈 칸 두 개 이상, 즉, space bar 키를 두번 누르면 줄넘김 효과가 됩니다.    
  
  
- 문단 제목  
  : 맨 왼쪽에 #을 하나만 넣으면 큰 제목이 됩니다.    
  : 맨 왼쪽에 ## 즉, #을 2개 연속으로 입력하면 중간제목이 됩니다.    
  : 맨 왼쪽에 ### 즉, #을 3개 연속으로 입력하면 작은 제목이 됩니다.     
        
        
* 글머리에 숫자를 넣으면 순번 글머리표(문단번호)가 됩니다.  
      
    
    
* 글머리에 "-"이나 "*"을 넣으면 점(●) 모양의 글머리표가 됩니다.

- 하이퍼링크는 다음과 같이 사용합니다. 더 자세한 문법도 아래 사이트를 참조 해주시면 됩니다.

## 4. 패키지 (Package)

현재 많이 사용되고 있는 주요 개발언어들은 빠른 개발과 코드 재사용성 등을 위해 객체지향 방식을 지원합니다.  
파이썬도 OS, 파일, Web, 암호화 등 다양한 기능 구현을 지원하기 위해 분야별로 주요 기능을 미리 만들어놓고 불러다 쓸 수 있습니다.  
이 묶음을 Library 또는 Module이라고 부르고, 이 보다 개념적으로 더 큰 묶음 단위는 Component라고 부르는데,   
파이썬에서는 이 단위를 Package라고 많이 부릅니다.


운영체제, 파일, 파일 형식, 압축, 암호화, 인터넷 등 부문별로 핵심적인 함수들은 파이썬에 내장되어 제공되기 때문에 내장 라이브러리 또는 표준 라이브러리라고 부르고 별도의 설치가 필요없이 불러오기(import)만 해도 쓸 수 있습니다.

표 형태의 데이터 처리에 용이한 Pandas,  
대표적인 시각화 패키지인 Matplotlib,   
수학/과학 모델링에 기반이 되는 SciPy,   
다양한 머신러닝 알고리즘을 지원하는 패키지인 Scikit-Learn 등이 Data 분석 및 시각화의 주요 패키지들 중 하나라고 볼 수 있습니다.

### 별도의 가상환경을 만들어서 geopandas 등의 패키지를 설치하고 싶을 때 - 이미 아나콘다를 쓰고 있는 경우

```python
# geo_env 라는 이름의 가상환경을 생성 
conda create -n geo_env 

# 아나콘다의 가상환경 리스트 조회 
conda env list

# 원하는 가상환경을 활성화 
conda activate geo_env

# 특정 가상환경이 활성화된 상태에서 주피터노트북을 설치
conda install jupyter notebook
conda install ipykernel

# 가상환경에 Kernel 연결(주피터노트북 연결)
python -m ipykernel install --user --name geo_env --display-name "geo_env"
# 주피터를 실행해서 보면 python3 밑에 geo_env가 보임. 클릭하면 geo_env 가상환경 상의 주피터로 접속됨 

#Kernel 연결 해제
#jupyter kernelspec unisnstall geo_env

# 특정 가상환경을 비활성화 
#conda deactivate geo_env

# 가상환경 삭제 
#conda remove -n geo_env --all 
```
